```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
train_folder = '../data/raw/train'
```

```python
dog_images = []
cat_images = []
```

```python
for filename in os.listdir(train_folder):
    # Ruta completa al archivo
    file_path = os.path.join(train_folder, filename)

    # Verificar que no sea un directorio
    if os.path.isfile(file_path):
        # Verificar si el archivo es de un perro o un gato
        if 'dog' in filename.lower():
            # Agregar el nombre del archivo a la lista de imágenes de perros
            dog_images.append(filename)
        elif 'cat' in filename.lower():
            # Agregar el nombre del archivo a la lista de imágenes de gatos
            cat_images.append(filename)
```

```python
print("Perros:", dog_images)
print("Gatos:", cat_images)
```

```python
top_4_dog_images = dog_images[:4]

# We load the selected dog_images.
plt.figure(figsize=(10, 10)) 

for i, dog_image in enumerate(top_4_dog_images):
    # We load the images
    img_path = os.path.join(train_folder, dog_image)
    img = Image.open(img_path)
    
    # We display the images.
    plt.subplot(2, 2, i + 1)
    plt.imshow(img)
    plt.title(dog_image)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
